```python
# duplicating the dataframes
arc_data_2 = arc_data.copy()
image_data_2 = image_data.copy()
tweets_data_2 = tweets_data.copy()
```

### Tidyness  Issues 1 ; rating_numeration and rating denominator should be calculated in same column and stored as float

### Define


#### The rating proportion will be calculated by dividing the numerator by denominated

### Code

```python
# divide the numerator by denomination and save the column as rating_proportion
rating =arc_data_2.rating_numerator/arc_data_2.rating_denominator
arc_data_2['rating'] =rating
```

```python
# drop the rating numerator and rating demominator column
arc_data_2.drop(columns=['rating_numerator','rating_denominator'],inplace = True)
```

### Test

```python
arc_data_2.head()
```

### Tidyness Issues 2 ; doggo,floofer, pupper,puppo column should be conbine in a single column

### Define

#### The dogo,floofer,pupper and puppo will be combine in a column

### Code

```python
# this code merge the three dog column into 1
breed_type = ['doggo','floofer','pupper','puppo']

arc_data_2[breed_type] = arc_data_2[breed_type].replace('None', np.nan)
def join_all(x):
    return ', '.join(x.dropna().astype(str))

arc_data_2['breed_type'] = arc_data_2[breed_type].apply(join_all, axis=1)

arc_data_2['breed_type']= arc_data_2['breed_type'].replace('', np.nan)
```

```python
# drop the columns that were use in merging but are no longer necessary
arc_data_2.drop(columns=['doggo','floofer','pupper','puppo'],inplace= True)
```

### Test

```python
# this output the value count of the new merge column 
arc_data_2['breed_type'].value_counts()
```

### Tidyness Issues 3 ; The image data and the tweet data should have being combined together with the archive data

### Define

#### The three dataset will be merge into 1 called the master Archive

### Code

```python
# merge the archive and the tweets data into twitter_archieve
twitter_archive= pd.merge(left=arc_data_2, right=tweets_data, left_on='tweet_id', right_on='id', how='inner')
```

```python
# twitter_archive merge with image data to get the master_data
master_archive= twitter_archive.merge(image_data, on='tweet_id', how='inner')
```
